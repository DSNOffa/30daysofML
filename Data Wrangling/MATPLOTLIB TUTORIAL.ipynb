## USING A SHORTHAND NOTATION
This is the format
fmt = '(colour)(marker)(line)'

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
 
# Notice the use of the shorthand format here
    
plt.plot(x,y, 'y^--' , label = '2x')

plt.plot(x,y)

plt.title('My First Graph!', fontdict = {'fontsize' : 20, 'fontname' : 'Times New Roman'})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

```python
plt.figure(figsize = (10,5))

plt.plot(gas.Year,gas.USA, 'b.-' ,label = 'USA')
plt.plot(gas.Year,gas.Canada, 'r.-', label = 'Canada')
plt.plot(gas.Year,gas['South Korea'], 'g.-', label = 'South Korea')
plt.plot(gas.Year,gas.Australia, 'y.-', label = 'Australia')

plt.title('Gas Prices over Time (in USD)', fontdict = {'fontweight': 'bold', 'fontsize': 20})

plt.xlabel('Years')
plt.ylabel('Gas prices (in USD)')

plt.xticks(gas.Year)

plt.legend()

plt.show()
```

```python
# Using a for loop to access all countries together
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker= '.')
        #print (country)
```

## Changing the location of the legend and placing in several places 

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker= '.', label = country)
        
        
plt.legend(loc = 'upper left')

plt.show()
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker= '.', label = country)
        
        
plt.legend(loc = 'best')

plt.show()
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker= '.', label = country)
        
        
plt.legend(loc = 'center left', bbox_to_anchor = (1.0,0.5))

plt.show()
```

```python
plt.figure(figsize = (10,5))

plt.plot(gas.Year,gas.USA, 'b.-' ,label = 'USA')
plt.plot(gas.Year,gas.Canada, 'r.-', label = 'Canada')
plt.plot(gas.Year,gas['South Korea'], 'g.-', label = 'South Korea')
plt.plot(gas.Year,gas.Australia, 'y.-', label = 'Australia')

plt.title('Gas Prices over Time (in USD)', fontdict = {'fontweight': 'bold', 'fontsize': 20})

plt.xlabel('Years')
plt.ylabel('Gas prices (in USD)')

plt.xticks(gas.Year)

plt.legend()

plt.savefig('Graph of Gas Prices in USD.png', dpi = 300)

plt.show() 
```
